The traditional wind rose plot plots wind speed and wind direction. The "Pollution Rose" plot in `vayu` has the same structure but replaces wind speed with pollutants.

```python
import pandas as pd
import matplotlib.pyplot as plt

from vayu import pollutionRose
```

Loading the data

```python
df = pd.read_csv("data/mydata.csv.tar.gz")
df.head()
```

Now, we plot the pollution rose for `pm25`, `pm10` and `o3`.

```python
pollutionRose(df, pollutant='pm25');
```

```python
pollutionRose(df, pollutant='pm10');
```

```python
pollutionRose(df, pollutant='o3');
```
